```python
############## Importing Necessary Libraries ##############
import pandas as pd
import requests, zipfile
from io import BytesIO
import os
import glob
import numpy as np
```

```python
############## Creating a List of date for which we want to download the data ##############
data_list = pd.DataFrame(pd.date_range(start='1/2/2021', end='5/1/2022', freq='M'), 
             columns=['Year_Month'])["Year_Month"].apply(lambda x : str(x).replace("-","")[:6]).tolist()
```

```python
############## Creating a Directory called Extracted_ZipFiles ##############
if "Extracted_ZipFiles" not in os.listdir():
    os.mkdir("Extracted_ZipFiles")
for year_month in data_list:
    ############## Defining the URL ##############
    url = f"https://s3.amazonaws.com/capitalbikeshare-data/{year_month}-capitalbikeshare-tripdata.zip"
    print(f"Starting Downloading datafile of {year_month}")
    ############## Splitting the URL to get the File ##############
    filename = url.split('/')[-1]
    ############## Downloading the File by sending the request to the URL ##############
    req = requests.get(url)
    print(f"Downloading the {year_month} file completed")
    ############## Extracting the ZipFiles contents ##############
    zip_ref = zipfile.ZipFile(BytesIO(req.content))
    print(f"Extracting the {year_month} file ")
    zip_ref.extractall(".\Extracted_ZipFiles")
    zip_ref.close()
```

```python
##### Creating a list of strings with files present in the Extracted_ZipFiles directory ##### 
file = os.path.join(".\Extracted_ZipFiles/*tripdata.csv")
files = glob.glob(file)

##### Lowercasing the Column names & replacing spaces with "_". Finally, replacing the original csv file with the new one.
for i in range(len(files)):
    df = pd.read_csv(files[i])
    df.columns = [str(col).lower().replace(" ","_") for col in pd.read_csv(files[i]).columns]
    df.to_csv(files[i], index=False)
    print(f"{files[i]} done!!!")
    print(" ")
```

```python
###### Checking whether all the Lowercasing of columns have been done or not ######
for i in range(len(files)):
    df = pd.read_csv(files[i])
    print(df.columns)
    print(" ")
```

```python
######## Creating a list of string with files of year 2020 ########
file_new = os.path.join(".\Extracted_ZipFiles/2020*tripdata.csv")
files_new = glob.glob(file_new)
print(f"Number of files of Year 2020 are - {len(files_new)}")
print(sorted(files_new)[3:])

######## Performing operations like renaming column names, changing the datatype of start_date & end_date columns, extracting duration of the ride and 
        # re-ording the columns ########
dataframes = []
for file in sorted(files_new)[3:]:
    x = pd.read_csv(file, usecols=['ride_id', 'started_at','ended_at', 'start_station_name', 'start_station_id','end_station_name', 'end_station_id', 'member_casual'])
    x.columns = ["bike_number", "start_date", "end_date", "start_station", "start_station_number", "end_station", "end_station_number", "member_type"]
    x["start_date"] = pd.to_datetime(x["start_date"])
    x["end_date"] = pd.to_datetime(x["end_date"])
    x["duration"] = (x['end_date'] - x['start_date'])/(np.timedelta64(1,"s"))
    x = x.loc[:,['duration', 'start_date', 'end_date', 'start_station_number','start_station', 'end_station_number', 'end_station', 'bike_number','member_type']]
    print(x.shape)
    dataframes.append(x)
```

```python
######### Concatenating the dataframes of the months starting from April till December of the Year 2020 #########
df_202004_202012 = pd.concat([dataframes[0],dataframes[1],dataframes[2],dataframes[3],dataframes[4],
                              dataframes[5],dataframes[6],dataframes[7],dataframes[8]], axis=0)
print(df_202004_202012.shape)
```
